```python
import requests
import json
from requests.auth import HTTPBasicAuth
import pandas as pd
from pathlib import Path
```

## Setting up the folder structure

```python
data_folder = Path.cwd().parent.joinpath('assets').joinpath('articles') 
print(data_folder)
```

```python
p = Path(data_folder).glob('**/*')
files = [x for x in p if x.is_file()]
```

## Parsing the json files

```python
def get_product_details_from_file(file):
    with open(str(file), 'r', encoding="utf-8") as f:        
        json_obj = json.loads(f.read())
        #print(json_obj)
        product={}
        product['id']= json_obj['id']
        product['product_name']= json_obj['name']
        try:
            product['origins'] = json_obj['origins']['producing_country']
        except:
            pass
        
        try:
            product['description'] = json_obj['description']['text']
        except:
            pass
        
        product['image_url'] = json_obj['image']['original']

        try:
            product['EAN'] = json_obj['links']['leshop']['app_url'].split('=')[-1]
        except:
            pass
        
        try:
            features = json_obj['features']
            #jprint(features)
            for feature in features:
                label_code = feature['label_code']
                label_code = label_code.lower()
                product[label_code] = feature['values'][0]['value']
        except:
            pass 
        
        try:
            categories = json_obj['categories']
            for category in categories:
                #print (category['name'], category['level'], category['code'])
                cat_level= 'cat_'+ str(category['level'])
                cat_level_name= 'cat_'+ str(category['level']) + '_name'
                product[cat_level] = category['code'] 
                product[cat_level_name] = category['name']
        except:
            pass
        print(product)
        return product
```

```python
len(files)
```

```python
product_data=[]
for i in range (len(files)):
    print(files[i])
    product = get_product_details_from_file(files[i])
    product_data.append(product)
```

## Reading product data into a pandas dataFrame

```python
product_df = pd.DataFrame(product_data)
product_df
```

## Reducing the number of columns for the database

```python
cols = (list(product_df.columns))
print(cols)
```

### Category columns

```python
cat_cols = [i for i in cols if i.startswith('cat_')]
print(cat_cols)
```

### All columns starting with 'miso_artvon' are required for article matching

```python
art_von_cols = [i for i in cols if i.startswith('miso_artvon')]
print(art_von_cols)
```

### Necessary data columns

```python
data_cols = ['id', 'product_name', 'EAN', 'origins']
```

```python
my_cols = data_cols + art_von_cols + cat_cols
print(my_cols)
```

## Create the reduced dataset with the minimized number of columns

```python
product_df_clean=product_df[my_cols]
product_df_clean
```

### Save the cleaned dataframe to csv to hand it over to the backend

```python
destination_folder = Path.cwd().parent.joinpath('assets')
```

```python
product_df_clean.to_csv(destination_folder.joinpath('products_clean.csv'),index=False)
```
